```python
import pandas as pd # importa a biblioteca
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('C:/Users/DSADM/Desktop/ATIV.-IABD/aula04/imigrantes_canada.csv')
```

```python
dados.tail()
```

```python
dados.head()
```

```python
dados.info()
```

```python
dados.set_index(['País'], inplace=True) #Seta a coluna pais como index
anos = list(map(str,range(1980,2013))) # transforma os anos em string ultilizado função map 
```

```python
anos
dados.tail(5)
```

```python
brasil = dados.loc['Brasil',anos] #filtro para pegar os dados somentee do brasil 
```

```python
brasil
```

```python
brasil_dic ={
    'ano' : brasil.index.tolist(), # pega os anos 
    'imigrantes' : brasil.values.tolist() # pega a quantidade de imigrantes 
}
```

```python
df_brasil = pd.DataFrame(brasil_dic) # criando o dataframe
```

```python
df_brasil.head() # exibe os imigrnates dp dataframe brasil
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks (['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks ([500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel ("anos",fontsize=14)
plt.title('Imigração do brasil para o Canadá', fontsize=18)
plt.ylabel ("Número de imigrantes",fontsize=14)
plt.show()
```

```python
argentina = dados.loc['Argentina',anos] #filtro para pegar os dados somente do argentina 
```

```python
argentina
```

```python
argentina_dic ={
    'ano' : argentina.index.tolist(), # pega os anos 
    'imigrantes' : argentina.values.tolist() # pega a quantidade de imigrantes 
}
```

```python
df_argentina = pd.DataFrame(argentina_dic) # criando o dataframe
```

```python
df_argentina.head() # exibe os imigrnates dp datafram argentina
```

```python
plt.plot(df_argentina['ano'],df_argentina['imigrantes'])
plt.xticks (['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks ([500,1000,1500,2000,2500,3000],fontsize=14)
plt.xlabel ("anos",fontsize=14)
plt.title('Imigração do argentina para o Canadá', fontsize=18)
plt.ylabel ("Número de imigrantes",fontsize=14)
plt.show()
```

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'],color = 'red' ,label = 'brasil')
plt.plot (df_argentina['ano'],df_argentina['imigrantes'],color = 'blue' ,label ='Argentina')
plt.xticks (['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks ([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.title('Imigrantes do brasil e argentina para o canadá',fontsize=14)
plt.ylabel("Nº de imigrantes",fontsize=18)
plt.xlabel('Anos',fontsize=14) # addd legenda ao gráfico
plt.legend()
plt.savefig ('Dados imigração.png') # salva 
plt.show()
```

```python
plt.bar(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks (['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.yticks ([0,500,1000,1500,2000,2500,3000],fontsize=14)
plt.title('Imigração do brasil para o canadá')
plt.show
```

```python
df_comparação =  dados.loc[['Brasil','Argentina'],anos]
```

```python
df_comparação
```

```python
df_comparação = df_comparação.T 
```

```python
df_comparação['T'] = df_comparação['Brasil'] + df_comparação['Argentina']
```

```python
plt.figure(figsize=(10,10))
plt.plot(df_comparação['Brasil'],label = 'Brasil', color= 'green')
plt.plot(df_comparação['Argentina'], label = 'Argentina',color = 'blue')
plt.plot(df_comparação ['T'],label= 'total',color = 'red')
plt.legend()
plt.xticks (['1980','1985','1990','1995','2000','2005','2010'],fontsize=14)
plt.title('imigrantes do brasil e \n argentina para o canadá', fontsize= 20)
plt.show

```
